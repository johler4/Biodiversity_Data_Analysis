```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Import the data

```python
observations = pd.read_csv('observations.csv')
print(observations.head())
print(observations.info())
```

```python
species = pd.read_csv('species_info.csv')
print(species.head())
print(species.info())
```

```python
# find the rows of 'species' where duplicated scientific names are listed:
# create a database of the scientific names
names = species['scientific_name']
# cycle through the list of scientific names to see which is duplicated, \
# select the rows that are associated with True values,
# sort the values based on 'scientific_name'
print(species[names.isin(names[names.duplicated()])].sort_values('scientific_name'))
```

> We can see that there are rows within our 'species' dataframe that have duplicate scientific names but different common names. We're going to assume that the experts who input this data knew what they were doing and that these are indeed different species from one another even though they share the same scientific name. This would explain why some species are listed in the 'observation' dataframe more than 4 times.

## Analyzing the Data:
**Questions/info to look into:**
1) What's the breakdown of the 'Endangered' species/categories?
    - We should be most concerned about these species.
2) What category of species appears most on the conservation list?
    - Is a specific category of species being affected more than others?
3) What are the counts of observations for conservation species in each park?
    - Which parks have lower numbers?
    - This could show that a specific park does not support the living habitat for one or more species.
4) What is some summary data for those species on each level of conservation?

```python
# Get list of endangered species only
endangered_species = species[species.conservation_status == 'Endangered']
print(endangered_species)
```

```python
sns.countplot(x='category', data=endangered_species)
plt.title('Count of Endangered Species by Category')
plt.xlabel('Category')
plt.show()
plt.clf()
```

> We can see that there are a total of 16 'endangered' species. 7 mammals, 4 birds, 3 fish, 1 vasular plant, and 1 amphibian.

```python
# find observation information for endangered species
endangered_species_observations = observations[observations.scientific_name.isin(endangered_species.scientific_name)]
print(endangered_species_observations)

# # prints a bar chart for each species count in each park:
# for species in endangered_species_observations['scientific_name']:
#     subset = endangered_species_observations[endangered_species_observations.scientific_name == species]
#     sns.barplot(x='scientific_name', y='observations', hue='park_name', data=subset)
#     plt.title(species + ' Count per Park')
#     plt.xlabel('Scientific Name')
#     plt.show()
#     plt.clf()
```

```python
endangered_species_observations.groupby(['park_name', 'scientific_name']).min()
```

```python
# What category of species appears most on the conservation list?
conservation_list_species = species[species.conservation_status.notnull()]
print(conservation_list_species.info())


sns.countplot(x='category', hue='conservation_status', data=conservation_list_species)
plt.xticks(rotation=30)
plt.show()
plt.clf()
```

```python
print(conservation_list_species.category.value_counts())
```

> Based on our data, we can see that there are a total of 191 species that lie somewhere on the conservation list. 16 of these species are listed as 'endangered' species: 7 mammals, 4 birds, 3 fish, 1 vasular plant, and 1 amphibian. The category with the highest majority of the conservation list is the bird category, followed by vasular plants and then mammals.

```python
print(species)
print(species[species.conservation_status.notnull()])
```

```python
# print total number of observations for each species that is on the conservation list
print(observations[observations.scientific_name.isin(conservation_list_species.scientific_name.unique())].groupby('park_name')['observations'].sum())
```

```python
# find observations of species that are on the conservation list:
print(observations[observations.scientific_name.isin(conservation_list_species.scientific_name)])
```

```python
# create a list of parks
parks = observations.park_name.unique()

for park in parks:
    plt.hist(observations.observations[observations.park_name == park], label=park, alpha=0.5)
    
plt.title('Observation Counts for Each Park')
plt.xlabel('# of observations')
plt.ylabel('# of times observation count occured')
plt.legend()
plt.show()
plt.clf()
```

```python
# find overall observation summary information only for conservation species at each park
print(observations[observations.scientific_name.isin(conservation_list_species.scientific_name)].groupby('park_name').describe())
```

> Viewing the data on observations of species that are on the conservation list, the lowest count, and lowest average count was at Great Smoky Mountains National Park. Without further investigation, this information could lead us to to believe that this state park is the least suitable habitat for the various species listed. Perhaps focusing actions to move animals/plants to other state parks would strengthen the chance of them surviving.
